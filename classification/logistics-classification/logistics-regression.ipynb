# Logistics Regression Classification

```python
#Import Library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Read Dataset
dataset = pd.read_csv("logit-classification.csv")
```

```python
dataset.head()
```

```python
#Userid, Gender is not more important
X = dataset.iloc[:,2:4].values
```

```python
X
```

```python
Y= dataset.iloc[:,-1].values
```

```python
Y
```

```python
#Split traning data and testing data as 80:20 ratios
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20, random_state=0)
```

```python
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
#Scaling done by StadardScaler range as -n to n
from sklearn.preprocessing import StandardScaler
sc  = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#make LogisticRegression model
from  sklearn.linear_model import LogisticRegression
classifer = LogisticRegression()
```

```python
#fitting traning data to model
classifer.fit(X_train, Y_train)
```

```python
#Predict value by model based on traning data for testing data
Y_pred = classifer.predict(X_test)
print('Predicted Value :', Y_pred)
```

```python

#printig metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("\n Accuracy ::", accuracy_score(Y_test, Y_pred))
print("\n Confusion Matrix ::\n", confusion_matrix(Y_test, Y_pred))
print("\n Classification Reports ::\n", classification_report(Y_test, Y_pred))
```

```python
#Heat Map for confusion Matrix
#plt.figure(6,4)
sns.heatmap(confusion_matrix(Y_test, Y_pred), annot=True, fmt="d", xticklabels=["Predicted 0", "Predicted 1"], yticklabels=["Actual 0", "Actual 1"])
plt.title("Confusion Matrix")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

```python
print("\n Classification Reports ::\n", classification_report(Y_test, Y_pred))
```

```python
#Bais, Variance score
bais = classifer.score(X_train, Y_train)
variance = classifer.score(X_test, Y_test)
```

```python
print("Bais :", bais)
print("Variance :", variance)
```
